```python
# separate into categorical (which are all objects now) and numerical variables to dummify categorical
fixed_obj = fixed.select_dtypes(include=object)
n_fixed_obj = n_fixed.select_dtypes(include=object)

fixed_num = fixed.select_dtypes(include=np.number)
n_fixed_num = n_fixed.select_dtypes(include=np.number)
```

```python
sns.heatmap(pd.DataFrame(fixed_num.corrwith(data['SalePrice']).abs().sort_values(ascending=False), columns=['Sale Price']))
```

```python
# fit on train and score on test and train
model_train_test = linreg.fit(X_train, y_train)
print(model_train_test.score(X_test, y_test), model_train_test.score(X_train, y_train))
```

```python
predictions_test = model_train_test.predict(X_test)
predictions_train = model_train_test.predict(X_train)
actuals_test = y_test
actuals_train = y_train
```

```python
residuals_test = y_test - predictions_test
residuals_train = y_train - predictions_train
```

## 2. Determine any value of *changeable* property characteristics unexplained by the *fixed* ones.

---

Now that you have a model that estimates the price of a house based on its static characteristics, we can move forward with part 2 and 3 of the plan: what are the costs/benefits of quality, condition, and renovations?

There are two specific requirements for these estimates:
1. The estimates of effects must be in terms of dollars added or subtracted from the house value. 
2. The effects must be on the variance in price remaining from the first model.

The residuals from the first model (training and testing) represent the variance in price unexplained by the fixed characteristics. Of that variance in price remaining, how much of it can be explained by the easy-to-change aspects of the property?

---

**Your goals:**
1. Evaluate the effect in dollars of the renovate-able features. 
- How would your company use this second model and its coefficients to determine whether they should buy a property or not? Explain how the company can use the two models you have built to determine if they can make money. 
- Investigate how much of the variance in price remaining is explained by these features.
- Do you trust your model? Should it be used to evaluate which properties to buy and fix up?

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
```

```python
# A:
# ss = StandardScaler()
# Xs = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
X = n_fixed_data
X_train = X[(fixed_data['YrSold']<2010)]
X_test = X[(fixed_data['YrSold']==2010)]
# Xs = pd.DataFrame(ss.fit_transform(X), columns=X.columns)
y_train = residuals_train
y_test = residuals_test
```

```python
optimal_lasso = LassoCV(n_alphas=200, cv=10)
optimal_lasso.fit(X_train, y_train)
print(optimal_lasso.alpha_)
```

```python
# use optimal lasso alpha and fit using pre-2010 data
lasso = Lasso(alpha = optimal_lasso.alpha_)
coefs = lasso.fit(X_train, y_train).coef_
```

```python
# score on 2010 predictors and 2010 residuals
print(lasso.score(X_test, y_test), lasso.score(X_train, y_train))
```

```python
plotdata = pd.DataFrame({'variable': X.columns, 'coefs': coefs, 'abs_coefs': np.abs(coefs)})
plotdata.set_index('variable', inplace=True)
```

```python
plotdata
```

```python
plot = plotdata.nlargest(30,'abs_coefs')
plot.plot(kind='barh', y='abs_coefs', figsize=(15,15))
```
